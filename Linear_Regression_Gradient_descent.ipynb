```python
#importing datasets from sklearn online source
import numpy as np
from sklearn import datasets, linear_model, metrics
diabetes = datasets.load_diabetes()
diabetes_X = diabetes.data 
#splitting data sets into two datasets
diabetes_X_train = diabetes_X[:-20]   #removes last twenty rows from dataset and stores in 'train' set.
diabetes_X_test = diabetes_X[-20:]    #takes only last twenty rows from the dataset and stores in 'test' set.
diabetes_y_train = diabetes.target[:-20]    #similarily with target output data
diabetes_y_test = diabetes.target[-20:]
X = diabetes_X_train
y = diabetes_y_train       #assigning them to variables
X1=diabetes_X_test
y1=diabetes_y_test
```

```python
yn=np.matrix(y)           #obtained values of y are in the form of simple scalar
ynew=yn.T                 #convert into martrix and made into form of a column matrix
ynew
```

```python
alpha=0.08
p=np.ones((422,1))          #a column vector consisting of m ones which are needed to be appended to X to make it m*n+1
new=np.append(p,X,axis=1)   #m is no.of rows(examples),n is no.of feautures(columns)
v=np.random.uniform(0,1,11) #initialising a solution dataset(theta) which is n+1*1
a=np.matrix(v)
w=a.T                       #making it into matrix of n+1*1 as the previous one is in scalar form.
n=422
for i in range(15000):
    h=np.matmul(new,w)      #h is matrix multiplication which gives m*1 matrix
    cost=(1/(2*n))*np.sum([val**2 for val in (h-ynew)])#cost function
    for j in range(11):
        a = (h-ynew).T
        xd=(1/n)*np.matmul(a,new[:,j]) #partial derivative
        w[j]=w[j]-alpha*xd #simultaneous update of theta
            
    print(cost)
    
print(w)
        
        
    
       
            
            
            
   
   

            
    

   
  
   

    





```
