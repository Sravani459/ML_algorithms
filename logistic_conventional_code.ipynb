```python
import tensorflow as tf
import numpy as np # linear algebra
import seaborn as sns #statistical data visualization
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
Iris=pd.read_csv('iris.csv')
Iris.head()
```

```python
Iris.shape
```

```python
Iris = Iris[:-50] 
Iris.shape
```

```python
sns.set(style='whitegrid')
sns.set(color_codes=True)
sns.lmplot(x="sepal_length",y="sepal_width", data=Iris, hue="species", fit_reg=False)
```

```python
Iris["species"].replace({"setosa": 0, "versicolor": 1}, inplace=True)
Iris.head() #replacing names with numericals
```

```python
X = Iris.drop('species', axis=1).values #feature matrix #input matrix formed after removing output column
y = Iris.species.values #output
res=np.matrix(y) #y is a scalar without dimensions
res1=res.T
Yout=res1    #making y into matrix and transpose to obtain as m*1 matrix 
print(Yout)
```

```python
train_index=np.random.choice(X.shape[0], 80, replace=False)  #generate random indices to split data
train_X=X[train_index,:]
print(train_X) #generate train and test input data
test_index= np.array(list(set(range(len(X))) - set(train_index)))
test_X=X[test_index,:]
print(test_X)
```

```python
train_Y=Yout[train_index] #split data into train_Y and test_Y
print(train_Y)
test_Y=Yout[test_index]
print(test_Y)
```

```python
alpha=0.08
p=np.ones((80,1))          #a column vector consisting of m ones which are needed to be appended to X to make it m*n+1
new=np.append(p,train_X,axis=1)   #m is no.of rows(examples),n is no.of feautures(columns)
v=np.random.uniform(0,1,5) #initialising a solution dataset(theta) which is 1*n+1
a=np.matrix(v)
w=a.T                       #making it into matrix of n+1*1 as the previous one is in scalar form.
m=80
cost_array=np.zeros(20000)
for i in range(20000):
    h=np.matmul(new,w)#h is matrix multiplication which gives m*1 matrix
    final_h=1/(1+np.exp(-h))#final_h is modified probability matrix when passed through sigmoid function
    temp=np.multiply(train_Y,np.log(final_h))
    temp1=np.multiply((1-train_Y),(np.log(1-final_h)))
    cost=(-1/m)*np.sum((temp+temp1)) #cost function is sum of two logarithmic functions
    for j in range(5):
        a = (final_h-train_Y).T
        xd=(1/m)*np.matmul(a,new[:,j]) #partial derivative
        w[j]=w[j]-alpha*xd #simultaneous update of theta
            
    print(cost)
    cost_array[i]=cost
    
print(w)
```

```python
epoch_array=np.arange(0,20000)
plt.plot(epoch_array,cost_array)
```

```python
p1=np.ones((20,1))          #a column vector consisting of m ones which are needed to be appended to X to make it m*n+1
new1=np.append(p1,test_X,axis=1)
r=np.matmul(new1,w)         #calculate the predicted value of output with obtained h values
hnew=1/(1+(np.exp(-r)))
hnew
```

```python
y_pred = np.zeros((20,1))
for i in range(20):
        if hnew[i][0] > 0.5:    #classify output based on sigmoid value,is positive if value>0.5
            y_pred[i][0] = 1
print(y_pred)   #output predicted is same as that of test output data given,hence modelling is true.
```
